```python
import pandas as pd
import numpy as np

#veriyükleme

csv_veriler = pd.read_csv("/content/drive/MyDrive/Colab Notebooks(ML)/BÖLÜM-3(PREDICTION)/veriler.csv")

print(csv_veriler)


# kategorik verilerin sayısal değerlere dönüştürülmesi(categoric -> numeric)

from sklearn import preprocessing

# ülke kategorik verisinin nümerik değere dönüşümü

ulkeler = csv_veriler.iloc[:, 0:1].values
#print(ulkeler)
print("")

# Label Encoder kullanarak verileri sayısal değerlere dönüştürme
le = preprocessing.LabelEncoder()
ulkeler[:, 0] = le.fit_transform(ulkeler[:,0])
'''
print("Label Encoder Kullanımı")
print(ulkeler)
print("")
'''
# One Hot Encoder kullanarak verileri sayısal değerlere döndürme
ohe = preprocessing.OneHotEncoder()
ulkeler = ohe.fit_transform(ulkeler).toarray()
'''
print("One Hot Encoder Kullanımı")
print(ulkeler)
print("")
'''

# cinsiyet kategorik verisinin nümerik değere dönüşümü

cinsiyet_numeric = csv_veriler.iloc[:, -1:].values
#print(cinsiyet_numeric)
print("")

# Label Encoder kullanarak verileri sayısal değerlere dönüştürme

le = preprocessing.LabelEncoder()
cinsiyet_numeric[:, -1] = le.fit_transform(cinsiyet_numeric[:,-1])
'''
print("Label Encoder Kullanımı (cinsiyet)")
print(cinsiyet_numeric)
print("")
'''

# One Hot Encoder kullanarak verileri sayısal değerlere döndürme

ohe = preprocessing.OneHotEncoder()
cinsiyet_numeric= ohe.fit_transform(cinsiyet_numeric).toarray()
'''
print("One Hot Encoder Kullanımı (cinsiyet)")
print(cinsiyet_numeric)
print("")
'''

# numpy dizilerinin dataframe'lere dönüştürlmesi

df_ulke = pd.DataFrame(data = ulkeler, index = range(22), columns = ["fr", "tr", "us"])
#print(df_ulke)

yas = csv_veriler.iloc[:, 1:4].values
df_boy_kilo_yas = pd.DataFrame(data = yas, index = range(22), columns = ["boy", "kilo", "yas"] )
#print(df_boy_kilo_yas)

cinsiyet_kategorik = csv_veriler.iloc[: , -1].values
df_cinsiyet = pd.DataFrame(data = cinsiyet_numeric[:,0:1], index = range(22), columns = ["cinsiyet"])
#print(df_cinsiyet)


# DATAFRAME BİRLEŞTİRME İŞLEMLERİ

# ülkeler ve boy kilo yaş sütunlarını içeren dataframe
df_1 = pd.concat([df_ulke, df_boy_kilo_yas], axis = 1)
df_1_deneme = df_1.iloc[:,0]
#print(df_1, "\n")

# ülkeler, boy, kilo, yaş sütunlarına ek cinsiyet sütununun eklendiği
# dataframe birleştirme işlemi
total_df = pd.concat([df_ulke, df_boy_kilo_yas, df_cinsiyet], axis=1)
print(total_df, "\n")



# Verilerin eğitim ve test için bölünme işlemlerinin gerçekleştirilimi

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df_1, df_cinsiyet, test_size=0.33, random_state=0)

print("X-TRAIN")
print(x_train, "\n")
print("X-TEST")
print(x_test, "\n")
print("Y-TRAIN")
print(y_train, "\n")
print("Y-TEST")
print(y_test, "\n")

# --TAHMİNLEME VE YAZDIRMA İŞLEMLERİ--

from sklearn.linear_model import LinearRegression

reg1 = LinearRegression()
reg1.fit(x_train, y_train)


# cinsiyet tahminlemesi

y_pred = reg1.predict(x_test)
'''
print("TEST KÜMESİ", "\n")
print(y_test, "\n")
print("PREDICTION", "\n")
print(y_pred, "\n")
'''

cinsiyet_pred = pd.DataFrame(data = y_pred, index = y_test.iloc[:,0].index, columns = ["prediction"])
#print(cinsiyet_pred, "\n")

cinsiyet_pred_y_test = pd.DataFrame(data = y_test, index = range(len(csv_veriler.iloc[:,0])), columns = ["cinsiyet"])
#print(cinsiyet_pred_y_test, "\n")

pred_result = pd.concat([cinsiyet_pred_y_test, cinsiyet_pred], axis = 1)

print(f"TEST-TAHMİN (%s) DATAFRAME" % (total_df.columns[-1]), "\n")
print(pred_result, "\n")


#################################################

# kilo tahminlemesi

reg2 = LinearRegression()
'''
kilo = total_df.iloc[:,4:5]
kilo_sol = total_df.iloc[:,:4]
kilo_sag = total_df.iloc[:,5:]

kilo_haric_veriler = pd.concat([kilo_sol, kilo_sag], axis = 1)

x_train, x_test, y_train, y_test = train_test_split(kilo_haric_veriler, kilo, test_size=0.33, random_state=0)
reg2.fit(x_train, y_train)

k_pred = reg2.predict(x_test)

print("TEST KÜMESİ", "\n")
print(y_test, "\n")
print("PREDICTION", "\n")
print(k_pred, "\n")


kilo_pred = pd.DataFrame(data = k_pred, index = y_test.iloc[:,0].index, columns = ["prediction"])
#print(kilo_pred, "\n")

kilo_pred_y_test = pd.DataFrame(data = y_test, index = range(len(csv_veriler.iloc[:,0])), columns = ["kilo"])
#print(kilo_pred_y_test, "\n")

pred_result = pd.concat([kilo_pred_y_test, kilo_pred], axis = 1)

print(f"TEST-TAHMİN (%s) DATAFRAME" % (total_df.columns[4]), "\n")
print(pred_result, "\n")

'''
#yas olmadan modelin kilo tahminlemesi


kilo = total_df.iloc[:,4:5]
kilo_sol = total_df.iloc[:,:4]
kilo_sag = total_df.iloc[:,6:]

kilo_haric_veriler = pd.concat([kilo_sol, kilo_sag], axis = 1)

x_train, x_test, y_train, y_test = train_test_split(kilo_haric_veriler, kilo, test_size=0.33, random_state=0)
reg2.fit(x_train, y_train)

k_pred = reg2.predict(x_test)

print("TEST KÜMESİ", "\n")
print(y_test, "\n")
print("PREDICTION", "\n")
print(k_pred, "\n")


kilo_pred = pd.DataFrame(data = k_pred, index = y_test.iloc[:,0].index, columns = ["prediction"])
#print(kilo_pred, "\n")

kilo_pred_y_test = pd.DataFrame(data = y_test, index = range(len(csv_veriler.iloc[:,0])), columns = ["kilo"])
#print(kilo_pred_y_test, "\n")

pred_result = pd.concat([kilo_pred_y_test, kilo_pred], axis = 1)

print(f"TEST-TAHMİN (%s) DATAFRAME" % (total_df.columns[4]), "\n")
print(pred_result, "\n")

###############################################################

# BACKWARD ELIMINATION  ile en uygun değişkenlerin seçimi

# p-value değeri < 0.05 olanlar bizim için anlamlı değişkenlerdir
# fr-tr-us-boy-yas-cinsiyet değişkenlerinden yas değişkeni
# 0.636 p-value ile bizim için çok anlam ifade etmeyen bir değişken
# bu yüzden backward elimination yöntemi ile bu değişkeni model eğitimine
# dahil etmiyoruz

import statsmodels.api as sm

# yaş dahil kilo tahminlemesi için regresyon analizi
'''
X = np.append(arr = np.ones((len(csv_veriler.iloc[:,0]),1)).astype(int), values = kilo_haric_veriler, axis = 1)
X_liste = kilo_haric_veriler.iloc[:,[0, 1, 2, 3, 4, 5]].values.astype(float)

# doğrusal regresyon analizi yapmak için kullanılan metot
model = sm.OLS(kilo, X_liste).fit()
print(model.summary())
'''

# yas hariç regresyon analizi
# yas değişkenini p-value değeri yüksek olduğundan model eğitiminden çıkardık
X = np.append(arr = np.ones((len(csv_veriler.iloc[:,0]),1)).astype(int), values = kilo_haric_veriler, axis = 1)
X_liste = kilo_haric_veriler.iloc[:,[0, 1, 2, 3, 4]].values.astype(float)

model = sm.OLS(kilo, X_liste).fit()
print(model.summary())

```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

#verilerin yklenmesi
veriler = pd.read_csv("/content/drive/MyDrive/Colab Notebooks(ML)/BOLUM-4(PREDICTION)/maaslar.csv")
print(veriler)

#dataframe slicing
x = veriler.iloc[:, 1:2]
X = x.values

#Numpy array dnm
y = veriler.iloc[:, 2:]
Y = y.values

#print(x)
#print(y)


#POLYNOMIAL REGRESSION

# 2. dereceden polinomiyal zellik ve grselletirmesi
poly_reg = PolynomialFeatures(degree = 2)
x_poly = poly_reg.fit_transform(X)
linear_reg = LinearRegression()
linear_reg.fit(x_poly, Y)
plt.scatter(X,Y, color = 'red')
plt.plot(X, linear_reg.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.show()
print()

# 2. dereceden polinoma gre eitim seviyesi 6 olan alann maa tahminlemesi
print(linear_reg.predict(poly_reg.fit_transform([[6]])))
print()


# 3. dereceden polinomiyal zellik ve grselletirmesi
poly_reg = PolynomialFeatures(degree = 3)
x_poly = poly_reg.fit_transform(X)
linear_reg = LinearRegression()
linear_reg.fit(x_poly, Y)
plt.scatter(X,Y, color = 'red')
plt.plot(X, linear_reg.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.show()
print()


# 3. dereceden polinoma gre eitim seviyesi 6 olan alann maa tahminlemesi
print(linear_reg.predict(poly_reg.fit_transform([[6]])))
print()



# 4. dereceden polinomiyal zellik ve grselletirmesi
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(X)
linear_reg = LinearRegression()
linear_reg.fit(x_poly, Y)
plt.scatter(X,Y, color = 'red')
plt.plot(X, linear_reg.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.show()
print()


# 4. dereceden polinoma gre eitim seviyesi 6 olan alann maa tahminlemesi
print(linear_reg.predict(poly_reg.fit_transform([[6]])))
print()



# rneklerde de grld zere polinomun derecesini artrdmzda overfitting olumaya balyor ve
# model test verisine ok fazla uyum salamaya balyor.
```

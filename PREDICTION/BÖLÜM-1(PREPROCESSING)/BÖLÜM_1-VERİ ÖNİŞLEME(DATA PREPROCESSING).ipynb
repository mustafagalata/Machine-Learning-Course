# **VERİ ÖNİŞLEME(DATA PREPROCESSING)**

```python
import pandas as pd

#veriyükleme

csv_veriler = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/eksikveriler.csv")

#print(csv_veriler)
print(csv_veriler[["boy", "yas", "kilo"]])

```

# EKSİK VERİLER

```python
import numpy as np
from sklearn.impute import SimpleImputer

#eksik veriler

imputer = SimpleImputer(missing_values=np.nan, strategy="mean")

yas = csv_veriler.iloc[:,1:4].values

#eğitim öncesi eksik verili kısım
print(yas)
print("")

# eğitim kısmı -> fit ile eksik verileri öğretiyoruz
imputer.fit(yas[:,1:4])

# fit ile öğrenilen parametreleri kullanarak veri setindeki eksik
# kısımları da transform ile dolduruyoruz
yas[:,1:4] = imputer.transform(yas[:,1:4])

print(yas)
print("")
```

# KATEGORİK VERİLERİN SAYISAL DEĞERLERE DÖNÜŞTÜRÜLMESİ


```python
# kategorik verilerin sayısal değerlere dönüştürülmesi(categoric -> numeric)

from sklearn import preprocessing

ulkeler = csv_veriler.iloc[:, 0:1].values
print(ulkeler)
print("")

# Label Encoder kullanarak verileri sayısal değerlere dönüştürme
print("Label Encoder Kullanımı")
le = preprocessing.LabelEncoder()
ulkeler[:, 0] = le.fit_transform(ulkeler[:,0])
print(ulkeler)
print("")

# One Hot Encoder kullanarak verileri sayısal değerlere döndürme
print("One Hot Encoder Kullanımı")
ohe = preprocessing.OneHotEncoder()
ulkeler = ohe.fit_transform(ulkeler).toarray()
print(ulkeler)
print("")
```

# VERİLERİN BİRLEŞTİRİLMESİ VE DATAFRAME OLUŞTURULMASI


```python
# numpy dizilerinin dataframe'lere dönüştürlmesi
df_ulke = pd.DataFrame(data = ulkeler, index = range(22), columns = ["fr", "tr", "us"])
print(df_ulke)
df_boy_kilo_yas = pd.DataFrame(data = yas, index = range(22), columns = ["boy", "kilo", "yas"] )
print(df_boy_kilo_yas)
cinsiyet = csv_veriler.iloc[: , -1].values
df_cinsiyet = pd.DataFrame(data = cinsiyet, index = range(22), columns = ["cinsiyet"])
print(df_cinsiyet)

# dataframe birlşetirme işlemleri

# ülkeler ve boy kilo yaş sütunlarını içeren dataframe
df_1 = pd.concat([df_ulke, df_boy_kilo_yas], axis = 1)
print(df_1)

# ülkeler, boy, kilo, yaş sütunlarına ek cinsiyet sütununun eklendiği
# dataframe birleştirme işlemi
total_df = pd.concat([df_ulke, df_boy_kilo_yas, df_cinsiyet], axis=1)
print(total_df)
```

# VERİ KÜMESİNİ EĞİTİM VE TEST OLARAK BÖLME

```python
from sklearn.model_selection import train_test_split

# Verilerin eğitim ve test için bölünme işlemlerinin gerçekleştirilimi

x_train, x_test, y_train, y_test = train_test_split(df_1, df_cinsiyet, test_size=0.33, random_state=0)

print(x_train)
print(x_test)
```

```python
from sklearn.preprocessing import StandardScaler

# Verilerin ölçeklendirilme işlemleri

S_sc = StandardScaler()
X_train = S_sc.fit_transform(x_train)
X_test = S_sc.fit_transform(x_test)

print(X_train)
print("")
print(X_test)
```

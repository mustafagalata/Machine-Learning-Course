```python
# KÜTÜPHANELERİN YÜKLENMESİ
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

# 1. VERİ YÜKLEME
veriler = pd.read_csv("/content/drive/MyDrive/Colab Notebooks(ML)/ÖDEV1(MULTIPLE LINEAR REGRESSION)/odev_tenis.csv")
#print(veriler, "\n")


# KATEGORİK VERİLERİN NÜMERİK VERİLERE DÖNÜŞTÜRÜLMESİ

le = LabelEncoder()
ohe = OneHotEncoder()

# windy ve play sütunlarının kategorik label ile encodelanması
le_veriler = veriler.apply(LabelEncoder().fit_transform)
#print(le_veriler, "\n")


#outlook sütununu one hot encoding ile encodelanması

ohe_veriler = ohe.fit_transform(le_veriler.iloc[:, :1]).toarray()
#print(ohe_veriler, "\n")


# NÜMERİK VERİLERİN DATAFRAME'E DÖNÜŞÜMÜ

df_outlook = pd.DataFrame(data = ohe_veriler, index = range(len(veriler)), columns = ["overcast", "rainy", "sunny"])
#print(df_outlook, "\n")

df_veriler = pd.DataFrame(data = le_veriler.iloc[:, 1:], index = range(len(veriler)), columns = ["temperature", "humidity", "windy", "play"])
#print(df_veriler, "\n")

df_son = pd.concat([df_outlook, df_veriler], axis = 1)
print(df_son, "\n")


# MODELİN TEST İÇİN BÖLÜNMESİ

x_train, x_test, y_train, y_test = train_test_split(df_son.iloc[:, :-1], df_son.iloc[:, -1:], test_size = 0.33, random_state = 0)

'''
print("x_train", "\n")
print(x_train, "\n")
print("x_test", "\n")
print(x_test, "\n")
print("y_train", "\n")
print(y_train, "\n")
print("y_test", "\n")
print(y_test, "\n")
'''

# MODELİN REGRESYONAL EĞİTİLMESİ VE TAHMİNLENMESİ
lr = LinearRegression()
lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)

# gerçek ve tahmin sonuçlarının kıyaslanması
df_kıyas = pd.DataFrame(data = y_test.values, index = y_test.index, columns = ["Gerçek Değerler"])
df_kıyas2 = pd.DataFrame(data = y_pred, index = y_test.iloc[:, 0].index, columns = ["Tahmin Edilen Değerler"])
df_kıyas_son = pd.concat([df_kıyas, df_kıyas2], axis = 1)
print(df_kıyas_son, "\n")


# MODELİN İSTATİSTİKSEL VERİLERİNİN ÇIKARILMASI

X = np.append(arr = np.ones((14, 1)).astype(int), values = df_son.iloc[:, :-1], axis = 1)
X_liste = df_son.iloc[:, [0, 1, 2, 3, 4, 5]].values.astype(float)

ols = sm.OLS(endog = df_son.iloc[:, -1], exog = X_liste).fit()
print(ols.summary())



```

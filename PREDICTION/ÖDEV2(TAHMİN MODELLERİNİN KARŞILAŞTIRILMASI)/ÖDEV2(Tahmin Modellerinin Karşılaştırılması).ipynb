```python
# Kütüphanelerin yüklenmesi
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import statsmodels.api as sm


#verilerin yüklenmesi
veriler = pd.read_csv("/content/drive/MyDrive/Colab Notebooks(ML)/ÖDEV2(TAHMİN MODELLERİNİN KARŞILAŞTIRILMASI)/maaslar_yeni.csv")

#verilerin yazdırılması
#print(veriler)

#"Calisan ID" sütununu çalışanların maaşlarıyla doğrudan bir bağlantısı olmaması, overfitting riskinin açığa çıkmaması için bağımsız
#değişkenler arasına almıyoruz. "unvan" sütununu ise dummy variable olmasından dolayı almıyoruz zaten "UnvanSeviyesi" sütunu bize
#çalışan tipi hakkında bilgi vermekte dolayısıyla "UnvanSeviyesi"ni kullanarak hem kategorik verinin nümerik veriye dönüşümü işlemine
#gerek kalmıyor hem de dummy variable kullanmanın da önüne geçmiş oluyoruz.


#bağımsız değişkenlerimiz(UnvanSeviyesi, Kidem, Puan)
x = veriler.iloc[:, 2:5]
X = x.values

#bağımlı değişkenimiz(Maas)
y = veriler.iloc[:, 5:]
Y = y.values

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X,Y)


model = sm.OLS(lin_reg.predict(X), X)
print(model.fit().summary())
```

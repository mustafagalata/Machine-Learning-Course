```python
#kütüphanelerin yüklenmesi
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

#verinin yüklenmesi
veriler = pd.read_csv("/content/drive/MyDrive/Colab Notebooks(ML)/CLASSIFICATION/1-LOGISTICREGRESSION/veriler.csv")

#bağımlı ve bağımsız değişkenlerin hazırlanması
x = veriler.iloc[:,1:4]
y = veriler.iloc[:,4:]
X = x.values
Y = y.values

#değişkenlerin eğitim ve test verilerine ayrıştırılması
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=0)


#değişkenlerin ölçeklendirilmesi
sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)


#model eğitimi ve tahminlemesi
log_reg = LogisticRegression(random_state=0)
log_reg.fit(X_train, y_train)

y_pred = log_reg.predict(X_test)
print(y_pred, "\n")
print(y_test, "\n")

#confusion matrix uygulaması
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)

```

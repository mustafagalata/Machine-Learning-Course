```python
#kütüphanelerin yüklenmesi
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

#verinin yüklenmesi

veriler = pd.read_csv('/content/drive/MyDrive/Colab Notebooks(ML)/CLUSTERING/musteriler.csv')

#veri önişleme
x = veriler.iloc[:, 3:5]
X = x.values

#model oluşturumu ve eğitimi
kmeans = KMeans(n_clusters= 3, init='k-means++', random_state = 0)
kmeans.fit(X)

#verilerin görselleştirimi
centers = kmeans.cluster_centers_
print(centers, '\n')
plt.scatter(x.iloc[:,0:1].values, x.iloc[:, 1:].values)
plt.scatter(centers[:, 0], centers[:, 1], marker='*', label='Küme Merkezleri')
plt.xlabel("İşlem Hacmi")
plt.ylabel("Maaş")
plt.title("Veri Noktalarının Maaş-İşlem Hacmi Dağılımı Grafiği")
plt.legend()
plt.show()
print()


#WCSS yöntemi ile dirsek noktası belirleyip küme sayısına karar verme

wcss_values_list = []

for i in range(0,10):
  kmeans = KMeans(n_clusters= i+1, init='k-means++', random_state = 0)
  kmeans.fit(X)
  wcss_values_list.append(kmeans.inertia_)

plt.plot(range(1,11),wcss_values_list, color='red')
plt.title("WCSS - Küme Sayısı Grafiği")
plt.xlabel("Küme Sayısı")
plt.ylabel("WCSS Değeri")
plt.show()




```

```python
#ktphanelerin yklenmesi
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

#verinin yklenmesi

veriler = pd.read_csv('/content/drive/MyDrive/Colab Notebooks(ML)/CLUSTERING/musteriler.csv')

#veri nileme
x = veriler.iloc[:, 3:5]
X = x.values

#model oluturumu ve eitimi
kmeans = KMeans(n_clusters= 3, init='k-means++', random_state = 0)
kmeans.fit(X)

#verilerin grselletirimi
centers = kmeans.cluster_centers_
print(centers, '\n')
plt.scatter(x.iloc[:,0:1].values, x.iloc[:, 1:].values)
plt.scatter(centers[:, 0], centers[:, 1], marker='*', label='Kme Merkezleri')
plt.xlabel("lem Hacmi")
plt.ylabel("Maa")
plt.title("Veri Noktalarnn Maa-lem Hacmi Dalm Grafii")
plt.legend()
plt.show()
print()


#WCSS yntemi ile dirsek noktas belirleyip kme saysna karar verme

wcss_values_list = []

for i in range(0,10):
  kmeans = KMeans(n_clusters= i+1, init='k-means++', random_state = 0)
  kmeans.fit(X)
  wcss_values_list.append(kmeans.inertia_)

plt.plot(range(1,11),wcss_values_list, color='red')
plt.title("WCSS - Kme Says Grafii")
plt.xlabel("Kme Says")
plt.ylabel("WCSS Deeri")
plt.show()




```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


veriler = pd.read_csv('/content/drive/MyDrive/Colab Notebooks(ML)/CLUSTERING/ASSOCIATION_RULE_MINING/APRIORI/sepet.csv', header = None)

transactions = []

for i in range(0, veriler.shape[0]):
  transactions.append([str(veriler.values[i,j]) for j in range(0, veriler.shape[1])])

from apyori import apriori

association_rules = apriori(transactions, min_support = 0.01, min_confidence = 0.2, min_lift = 3.0, min_length = 2)

print(list(association_rules))
```

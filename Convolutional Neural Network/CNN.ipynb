```python
# Kütüphanelerin yüklenmesi
from keras.models import Sequential
from keras.layers import Convolution2D, MaxPooling2D, Flatten, Dense
```

```python
# İlkleme
classifier = Sequential()

# Adım 1 - Convolution
classifier.add(Convolution2D(32, 3, 3, input_shape= (64, 64, 3), activation='relu'))

# Adım 2 - Pooling
classifier.add(MaxPooling2D(pool_size=(2, 2)))

# 2.Convolution katmanı
classifier.add(Convolution2D(32, 3, 3, activation='relu'))

# 2.Pooling katmanı
classifier.add(MaxPooling2D(pool_size=(2, 2)))

# Adım 3 - Flattening
'''
Özellik çıkardığımız resmi tek boyutlu bir vektöre dönüştürüyoruz
yapay sinir ağına vereceğimiz için ön işleme olarak bunu yapıyoruz)
'''
classifier.add(Flatten())

# Adım 4 - Yapay Sinir Ağı
classifier.add(Dense(units = 128, activation='relu'))
classifier.add(Dense(units = 1, activation='sigmoid'))

# Convolutional Neural Network
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])


# CNN ve Resimler

from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_data_generator = ImageDataGenerator(rescale = 1./255,
                                          shear_range = 0.2,
                                          zoom_range = 0.2,
                                          horizontal_flip = True)

test_data_generator = ImageDataGenerator(rescale = 1./255,
                                         shear_range = 0.2,
                                         zoom_range = 0.2,
                                         horizontal_flip = True)

training_set = train_data_generator.flow_from_directory('veriler/training_set',
                                                        target_size = (64, 64),
                                                        batch_size = 32,
                                                        class_mode = 'binary')

test_set = train_data_generator.flow_from_directory('veriler/test_set',
                                                    target_size = (64, 64),
                                                    batch_size =  32,
                                                    class_mode = 'binary')

steps_per_epoch_variable = training_set.samples // training_set.batch_size
validation_steps_variable = test_set.samples // test_set.batch_size
total_epochs = 50

# Manuel epoch döngüsünü başlatıyoruz
for epoch in range(total_epochs):
    print(f"--- Epoch {epoch + 1}/{total_epochs} Başlıyor ---")

    history = classifier.fit(training_set,
                             epochs=1,
                             validation_data=test_set,
                             steps_per_epoch=steps_per_epoch_variable,
                             validation_steps=validation_steps_variable,
                             verbose=1)

    print(f"--- Epoch {epoch + 1}/{total_epochs} Tamamlandı ---\n")

print("Tüm eğitim süreci bitti.")

import numpy as np
import pandas as pd


test_set.reset()
prediction = classifier.predict(test_set, verbose = 1)
prediction[prediction > .5] = 1
prediction[prediction <= .5] = 0

print("Tahmin gerçekleşti.")


test_labels = []

for i in range(0, len(test_set)):
  test_labels.extend(np.array(test_set[i][1]))

print(test_labels)


from sklearn.metrics import confusion_matrix

karmasiklik_matrisi = confusion_matrix(test_labels, prediction)
print(karmasiklik_matrisi)

# Tahminlerin dataframe olarak gösterilmesi
dosya_isimleri = test_set.filenames
sonuc = pd.DataFrame()
sonuc['Dosya_İsimleri'] = dosya_isimleri
sonuc['Tahminler'] = prediction.flatten()
sonuc['Gercek Değerler'] = test_labels

print(sonuc)
```
